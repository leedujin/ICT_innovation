```python
input_data = np.array([1, 2])
target_data = np.array([1])

learning_rate = 1e-1

def numerical_derivative(f, x):
  it = np.nditer(x, flags = ['multi_index'], op_flags=['readwrite'])
  grad = np.zeros_like(x)

  while not it.finished:
    delta_x = 1e-4
    idx = it.multi_index
    tmp_value = x[idx] 
    
    x[idx] = float(x[idx]) + delta_x
    fx1 = f(x)

    x[idx] = tmp_value
    x[idx] = float(x[idx]) - delta_x
    fx2 = f(x)

    grad[idx] = fx1 - fx2 / 2 * delta_x
    x[idx] = tmp_value

    it.iternext

  return grad

def sigmoid(x):
  return 1/ (1 + np.exp(-x))
```

```python
input_nodes = 2
hidden_nodes = 3
output_nodes = 1
```

```python
W2 = np.random.rand(input_nodes, hidden_nodes)
b2 = np.random.rand(hidden_nodes,)
W3 = np.random.rand(hidden_nodes, output_nodes)
b3 = np.random.rand(output_nodes,)
```

```python

def feedforward(xdata):

  Z2 = np.dot(xdata, W2) + b2
  A2 = sigmoid(Z2)

  Z3 = np.dot(A2, W3) + b3
  A3 = sigmoid(Z3)

  Y = A3

  loss = np.sum((target_data - Y)**2) / len(Y)

  return loss

feedforward(input_data)
```

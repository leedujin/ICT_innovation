```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
gen = ImageDataGenerator(rescale = 1./255)
```

```python
import zipfile

with zipfile.ZipFile('/content/test_class_mode.zip') as target_file:
  target_file.extractall('test_class_mode')
```

```python
data_path = '/content/test_class_mode/'

batch_size = 2

data_gen = gen.flow_from_directory(directory = data_path,
                                   batch_size = batch_size,
                                   shuffle = True,
                                   target_size = (100, 100),
                                   class_mode = 'binary')


print(data_gen.class_indices)
print(data_gen.num_classes)
print(data_gen.class_mode)
```

```python
img, label = data_gen.next()

for i in range(len(label)):
  print('label = ', label[i])

plt.figure(figsize = (8, 8))

for i in range(len(img)):

  plt.subplot(1, len(img), i+1)
  plt.xticks([])
  plt.yticks([])
  plt.title(str(label[i]))

  plt.imshow(img[i])

plt.show()
```

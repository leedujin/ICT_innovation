```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.datasets import mnist
```

```python
(x_train, t_train), (x_test, t_test) = mnist.load_data()
print(x_train.shape, t_train.shape, x_test.shape, t_test.shape)
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0

print(x_train.max(), x_train.min())
print(x_test.max(), x_test.min())
```

```python
model = Sequential()

model.add(Conv2D(input_shape = (28, 28, 1),filters = 32, kernel_size = 5, padding = 'SAME', activation = 'relu', use_bias = True))

model.add(MaxPool2D(pool_size = (2, 2), padding = 'SAME'))

model.add(Flatten())

model.add(Dense(10, activation = 'softmax'))
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = Adam(learning_rate = 0.001), metrics = ['accuracy'])
```

```python
model.summary()
```

```python
hist = model.fit(x_train.reshape(-1, 28, 28, 1), t_train, batch_size = 50, epochs = 50, validation_split = 0.2)
```

```python
import matplotlib.pyplot as plt

plt.title('Loss')
plt.xlabel('epochs')
plt.ylabel('loss')

plt.plot(hist.history['loss'], label = 'training loss')
plt.plot(hist.history['val_loss'], label = 'validation loss')
plt.legend(loc = 'best')

plt.show()
```

```python
plt.title('Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.plot(hist.history['accuracy'], label = 'training accuracy')
plt.plot(hist.history['val_accuracy'], label = 'validation accuracy')
plt.legend(loc = 'best')

plt.show()
```

```python
ret_val = model.predict(x_test.reshape(-1, 28, 28, 1))

print('type(ret_val)')
```
